```python
import pandas as pd
from os import path

import glob

dfs = []

DATA_DIR = path.join(path.pardir, path.pardir, 'data', 'microbenchmark_primitives', 'throughput')


for name in glob.glob(f'{DATA_DIR}/dynamodb_*.csv'):
    print(name)
    n = name[name.find('dynamodb_'):]
    n = n.split('_')
    workers = int(n[1])
    rps = int(n[2])
    idx = int(n[3].split('.')[0])
    gen = pd.read_csv(name)
    
    if idx != workers - 1:
        gen['rps'] = rps
    dfs.append(gen)


for name in glob.glob(f'{DATA_DIR}/lock_*.csv'):
    print(name)
    n = name[name.find('lock_'):]
    n = n.split('_')
    workers = int(n[1])
    rps = int(n[2])
    idx = int(n[3].split('.')[0])
    gen = pd.read_csv(name)
    
    if idx != workers - 1:
        gen['rps'] = rps
    dfs.append(gen)

    
df = pd.concat(dfs)
print(df.tail())
df.drop(df.loc[df['end'] < 0].index, inplace=True)
df['begin_timestamp'] = pd.to_datetime(df['begin'], unit='s')
df['end_timestamp'] = pd.to_datetime(df['end'], unit='s')
df.set_index('end_timestamp', inplace=True)
print(df)
df['count'] = 1
```

```python
data = []
for group_name, df_group in df.groupby(['workers']):
    #print(group_name)
    for group_name2, df_group2 in df_group.groupby(['rps']):
        #print(group_name2)
        group_sorted = df_group2.sort_index()
        #print(group_name, group_name2, df_group2)
        #df_group2.to_csv('file')
        #group_sorted.to_csv('file2')
        rolling = group_sorted.rolling('1s')['count'].sum()
        data.append([(group_name - 1)*120 + group_name2, rolling.mean(), group_name, 'mean'])
        data.append([(group_name - 1)*120 + group_name2, rolling.median(), group_name, 'median'])
        data.append([(group_name - 1)*120 + group_name2, rolling.quantile(.99), group_name, 'p99'])
        data.append([(group_name - 1)*120 + group_name2, rolling.max(), group_name, 'max'])
print(data)
processed_windows = pd.DataFrame(data, columns=['rps', 'val', 'workers', 'type'])
processed_windows['benchmark'] = 'dynamodb'
#diff = df_group['timestamp'].max() - df_group['timestamp'].min()
#count = df_group['count'].sum()
##print(count/diff)
#print(rolling.mean())
#break
```
