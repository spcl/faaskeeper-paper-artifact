```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import ticker
from scipy import stats
from os.path import join, pardir

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

#sns.set(rc={'figure.figsize':(8,5)})
sns.set(rc={'figure.figsize':(8,5)})
sns.set_style("whitegrid")

fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(2,1)
aws = fig3.add_subplot(gs[0, 0])
gcp = fig3.add_subplot(gs[1, 0])

DATA_DIR = join(pardir, pardir, 'data', 'read', 'gcp')
sns.set(rc={'figure.figsize':(8,7)})
sns.set_style("whitegrid")

def p50_estimator(x):                                            
    return np.percentile(x, 50)       
def p95_estimator(x):                                            
    return np.percentile(x, 95)         
       

# DynamoDB
key_value = pd.read_csv(join(DATA_DIR, 'datastore.csv'), index_col=0)
#redis = pd.read_csv(join(DATA_DIR, 'redis.csv'), index_col=0)
persistent = pd.read_csv(join(DATA_DIR, 'cloud_storage.csv'), index_col=0)
zookeeper = pd.read_csv(join(pardir, pardir, 'data', 'read', 'zookeeper', 'gcp', 'interzone.csv'), index_col=0)


key_value['data'] /= 1000.0
#redis['data'] /= 1000.0
persistent['data'] /= 1000.0
zookeeper['data'] /= 1000.0
key_value['size'] /= 1024.0
#redis['size'] /= 1024.0
persistent['size'] /= 1024.0
zookeeper['size'] /= 1024.0

ax=gcp
sns.lineplot(
    data=key_value.loc[key_value['op'] == 'read'], x='size', y='data', color='r',
    estimator=p50_estimator, label='FaaSKeeper, Datastore',ax=ax)

ax = sns.lineplot(data=persistent.loc[persistent['op'] == 'read'], x='size', y='data',
                  estimator=p50_estimator, label='FaaSKeeper, Cloud Storage',ax=ax)#,fontsize=20)

sns.lineplot(data=zookeeper.loc[zookeeper['op'] == 'read'], x='size', y='data',
             #estimator=p50_estimator, label='ZooKeeper (Self-Hosted)',ax=ax)#,fontsize=20)
             estimator=p50_estimator, label='ZooKeeper',ax=ax)#,fontsize=20)

ax.set_xlabel('Size [kB]',fontsize=12)
ax.set_ylabel('Time [ms]',fontsize=12)
ax.set_xlim(0, 400)
ax.set_ylim(0, 60)
ax.set_title("Google Cloud Platorm (GCP)", fontweight='bold', fontsize=16)



ax.annotate(
    'FaaSKeeper with GCP Datastore vs AWS DynamoDB:\n2.38x slower on small nodes, 30% faster on large nodes.',
    xy=(10, 10), xytext=(40,20), fontsize=12, ha='left', fontweight='bold',
    arrowprops=dict(facecolor='black', shrink=0.05)
)
ax.annotate(
    '',
    xy=(400, 15), xytext=(320,30), fontsize=12, ha='left', fontweight='bold',
    arrowprops=dict(facecolor='black', shrink=0.05)
)

ax.annotate('Object storage slower than AWS S3.',
            xy=(125, 55), xytext=(50, 40), fontsize=12, fontweight='bold',
            arrowprops=dict(facecolor='black', shrink=0.05)
)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.31), ncol=3)

DATA_DIR = join(pardir, pardir, 'data', 'read')
ax=aws

key_value = pd.read_csv(join(DATA_DIR, 'key_value.csv'), index_col=0)
redis = pd.read_csv(join(DATA_DIR, 'redis.csv'), index_col=0)
persistent = pd.read_csv(join(DATA_DIR, 'persistent.csv'), index_col=0)
zookeeper = pd.read_csv(join(DATA_DIR, 'zookeeper', 'intrazone.csv'), index_col=0)
key_value['data'] /= 1000.0
redis['data'] /= 1000.0
persistent['data'] /= 1000.0
zookeeper['data'] /= 1000.0
key_value['size'] /= 1024.0
redis['size'] /= 1024.0
persistent['size'] /= 1024.0
zookeeper['size'] /= 1024.0
key_value = key_value.drop(key_value[key_value['size'] > 256].index)
persistent = persistent.drop(persistent[persistent['size'] > 256].index)
zookeeper = zookeeper.drop(zookeeper[zookeeper['size'] > 256].index)

sns.lineplot(
    data=key_value.loc[key_value['op'] == 'write'], x='size', y='data', color='r',
    estimator=p50_estimator, label='FaaSKeeper, DynamoDB',ax=ax)#,fontsize=20)

sns.lineplot(
    data=redis.loc[redis['op'] == 'read'], x='size', y='data', color='g',
    estimator=p50_estimator, label='FaaSKeeper, Redis',ax=ax)#,fontsize=20)

ax = sns.lineplot(data=persistent.loc[persistent['op'] == 'read'], x='size', y='data',
                  estimator=p50_estimator, label='FaaSKeeper, S3',ax=ax)#,fontsize=20)
sns.lineplot(data=zookeeper.loc[zookeeper['op'] == 'read'], x='size', y='data',
             #estimator=p50_estimator, label='ZooKeeper (Self-Hosted)',ax=ax)#,fontsize=20)
             estimator=p50_estimator, label='ZooKeeper',ax=ax)#,fontsize=20)


ax.set_xlabel('Size [kB]',fontsize=12, labelpad=0)
ax.set_ylabel('Time [ms]',fontsize=12)

ax.annotate('FaaSKeeper: cloud-native storage\ndominates read time.',
            xy=(250, 10), xytext=(20,8), fontsize=12, fontweight='bold',
            arrowprops=dict(facecolor='black', shrink=0.05)
)
ax.annotate('',
            xy=(250, 15), xytext=(200,12.5), fontsize=12, fontweight='bold',
            arrowprops=dict(facecolor='black', shrink=0.05)
)

ax.annotate('FaaSKeeper with in-memory cache\non par with self-hosted ZooKeeper.',
            xy=(250, 1.5), xytext=(50,1.8), fontsize=12, fontweight='bold'
            #arrowprops=dict(facecolor='black', shrink=0.05)
)
ax.annotate('',
            xy=(250, 1.5), xytext=(200,2.5), fontsize=12, fontweight='bold',
            arrowprops=dict(facecolor='black', shrink=0.05)
)
ax.set_title("Amazon Web Services (AWS)", fontweight='bold', fontsize=16)

#ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25), ncol=2)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.28), ncol=4,fontsize=10)
#plt.rcParams['figure.dpi'] = 500
#plt.tight_layout()

plt.savefig("performance_plot_combined.pdf",bbox_inches='tight',pad_inches = 0, transparent=False)
```
